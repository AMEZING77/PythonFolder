- 常用的绘图操作
  - 使用np绘制二维数组Blank
  - 改变np数组的颜色
  - rectangle 绘制矩形
  - cirlce 绘制圆形
  - line 绘制线段
  - putText 绘制文字



```python
# draw
import cv2 as cv
import numpy as np

blank = np.zeros((500, 500, 3), dtype="uint8")
# cv.imshow("Blank", blank)


# change color
rowMin = 100
rowMax = 200
colMin = 300
colMax = 400
blank_red = np.zeros((500, 500, 3), dtype="uint8")
blank_red[rowMin:rowMax, colMin:colMax] = [255, 0, 0]
cv.imshow("Red", blank_red)

# draw a rectangle
cv.rectangle(
    blank_red,
    (colMin, rowMin),
    (colMax, rowMax),
    (0, 255, 0),
    thickness=3,
    # fill the rectangle
    # thickness=-1,
)
cv.imshow("Rectangle", blank_red)

# draw a circle
pos = blank_red.shape[1] // 2, blank_red.shape[0] // 2
cv.circle(blank_red, pos, 40, (0, 0, 255), thickness=3)
cv.imshow("Circle", blank_red)

# draw a line
cv.line(blank_red, (0, 0), (500, 500), (255, 255, 255), thickness=3)
cv.imshow("Line", blank_red)

# draw a text
cv.putText(
    blank_red,
    "Hello",
    (100, 100),
    cv.FONT_HERSHEY_TRIPLEX,
    1.0,
    (0, 100, 0),
    thickness=3,
)
cv.imshow("Text", blank_red)

cv.waitKey(0)
cv.destroyAllWindows()
```
