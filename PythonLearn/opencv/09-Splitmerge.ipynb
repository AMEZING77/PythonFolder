- 三色分离BGR `cv.split`
  - 当前通道约接近白色，说明该像素在该通道上亮度较高，反之则较低。
  - 三色图的Shape大小都是一致的
- 合并BGR `cv.merge([b,g,r])`
- 结合空白背景显示



```python
import cv2 as cv
import numpy as np

img = cv.imread("Photos/4.png")
# cv.imshow("image", img)
# resize 缩放
resized = cv.resize(img, None, fx=0.2, fy=0.2)
cv.imshow("Resized", resized)
img = resized

blank = np.zeros(img.shape[0:2], np.uint8)

# 三原色分离
b, g, r = cv.split(img)
blue = cv.merge([b, blank, blank])
green = cv.merge([blank, g, blank])
red = cv.merge([blank, blank, r])

cv.imshow("BlueBlank", blue)
cv.imshow("GreenBlank", green)
cv.imshow("RedBlank", red)

cv.imshow("Blue", b)
cv.imshow("Green", g)
cv.imshow("Red", r)

print(img.shape)
print(b.shape)
print(g.shape)
print(r.shape)

# 合并三色图
merged = cv.merge([b, g, r])
cv.imshow("Merged", merged)


cv.waitKey(0)
cv.destroyAllWindows()
```
