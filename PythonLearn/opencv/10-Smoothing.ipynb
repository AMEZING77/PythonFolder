- 平滑滤波方法介绍：
>

- 选择Kernel row*col
- 平均模糊法 将周围像素的平均值赋给中心像素 `cv.blur(img, (row, col))`
- 高斯模糊法 高斯核卷积 `cv.GaussianBlur(img, (row, col), 0)` 第三个参数为标准差，默认为0，此时自动计算为row/col*0.35
- 中值模糊法 将周围像素的中位数赋给中心像素 `cv.medianBlur(img, kernel_size)`
- bilateral 双边滤波 双边滤波在保持边缘清晰的同时模糊了中间的像素点，适用于去噪同时保留边缘信息 `cv.bilateralFilter(img, 直径, sigmaColor, SigmaSpace)`
  





```python
import cv2 as cv
import numpy as np

img = cv.imread("Photos/3.png")
# cv.imshow("image", img)
# resize 缩放
resized = cv.resize(img, None, fx=0.2, fy=0.2)
cv.imshow("Resized", resized)
img = resized

# Averaging
averaged = cv.blur(img, (3, 3))
cv.imshow("Averaging", averaged)

# Gaussian Blur
gauss = cv.GaussianBlur(img, (7, 7), 0)
cv.imshow("Gaussian Blur", gauss)

# Median Blur
median = cv.medianBlur(img, 7)
cv.imshow("Median Blur", median)

# Bilateral Filter
bilateral = cv.bilateralFilter(img, 9, 75, 75)
cv.imshow("Bilateral Filter", bilateral)


cv.waitKey(0)
cv.destroyAllWindows()
```
