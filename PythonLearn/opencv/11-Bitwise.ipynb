- bitwise介绍
  - 掩码：一个与原图像大小相同的矩阵，其中的元素只能是0或1
  - And 返回图像的像素值与掩码的对应位都为1时，该位置才为1
  - OR 返回图像的像素值与掩码的对应位都为0时，该位置才为0
  - XOR 返回图像的像素值与掩码的对应位不同时，该位置才为1
  - NOT 返回图像的像素值与掩码的对应位不同时，该位置才为1



```python
import cv2 as cv
import numpy as np

blank = np.zeros((400, 400), dtype=np.uint8)

rectangle = cv.rectangle(blank.copy(), (30, 30), (370, 370), 255, thickness=cv.FILLED)
circle = cv.circle(blank.copy(), (200, 200), 200, 255, thickness=cv.FILLED)

cv.imshow("Rectangle", rectangle)
cv.imshow("Circle", circle)


# bitwise AND 返回两个图像的重叠部分
bitwise_and = cv.bitwise_and(rectangle, circle)
cv.imshow("Bitwise AND", bitwise_and)

# bitwise OR 返回两个图像的全部部分，包括重叠的部分
bitwise_or = cv.bitwise_or(rectangle, circle)
cv.imshow("Bitwise OR", bitwise_or)

# bitwise XOR 返回两个图像的非重叠部分，但不包括重叠的部分
bitwise_xor = cv.bitwise_xor(rectangle, circle)
cv.imshow("Bitwise XOR", bitwise_xor)

# bitwise NOT 返回图像的取反，即非重叠部分变为白色，重叠部分变为黑色
bitwise_not = cv.bitwise_not(circle)
cv.imshow("Bitwise NOT", bitwise_not)

cv.waitKey(0)
cv.destroyAllWindows()
```
