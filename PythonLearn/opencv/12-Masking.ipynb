- masking 进行图像的遮罩处理，比如在图像上绘制一个圆形遮罩
- `cv.bitwise_and(img,img,mask)` 进行图像的与操作，可以用来实现遮罩效果



```python
import cv2 as cv
import numpy as np

img = cv.imread("Photos/3.png")
# cv.imshow("image", img)
# resize 缩放
resized = cv.resize(img, None, fx=0.2, fy=0.2)
cv.imshow("Resized", resized)
img = resized

blank = np.zeros(img.shape[0:2], dtype=np.uint8)
cv.imshow("Blank", blank)

circle = cv.circle(
    blank.copy(), (img.shape[1] // 2, img.shape[0] // 2), 100, 255, thickness=cv.FILLED
)
cv.imshow("MaskCircle", circle)

rectangle = cv.rectangle(blank.copy(), (30, 30), (200, 200), 255, thickness=cv.FILLED)
cv.imshow("MaskRectangle", rectangle)

masked_Circle = cv.bitwise_and(img, img, mask=circle)
cv.imshow("Masked", masked_Circle)

mask_Rectangle = cv.bitwise_and(img, img, mask=rectangle)
cv.imshow("MaskedRectangle", mask_Rectangle)

cv.waitKey(0)
cv.destroyAllWindows()
```
